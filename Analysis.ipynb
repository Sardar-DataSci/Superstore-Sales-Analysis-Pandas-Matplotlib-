🎯 Project Objective

Analyze a store’s sales data to answer business questions:

What is the total revenue?

Which category sells the most?

Which region performs best?

How do monthly sales trends look?

```python
import pandas as pd

data = {
    "Month": ["Jan","Feb","Mar","Apr","May","Jun"],
    "Sales": [12000,15000,18000,16000,20000,22000],
    "Orders": [120,150,180,160,200,220],
    "Category": ["Electronics","Clothing","Electronics","Home","Clothing","Electronics"],
    "Region": ["North","South","East","West","North","South"]
}

df = pd.DataFrame(data)
df.to_csv("Sales Dataset.csv")

print(df)
```

Basic Data Explotataion

```python
df.head() #use to get basic first info
df.tail()
df.info() #dataset information
df.describe() #stastics
```

Monthly Sales Trend. Visulaization by using matplotlib library

```python
import matplotlib.pyplot as plt

plt.plot(df["Month"],df["Sales"],color="red",marker="o")
plt.tight_layout()
plt.title("Monthly Sales Trend",color="Blue",fontsize=20,fontweight="bold")
plt.xlabel("Months",color="Gray",fontsize=15,labelpad=20)
plt.ylabel("Slaes",color="Gray",fontsize=15,labelpad=20)
plt.grid(True)
plt.style.use("fivethirtyeight")
plt.show()
```

Sales By Region

```python
Sales_by_Region=df.groupby("Region")["Sales"].sum()
plt.plot(Sales_by_Region.index,Sales_by_Region.values)
plt.title("Sales By Region")
plt.show()
```
